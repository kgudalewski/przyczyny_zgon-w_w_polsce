```python
import pandas as pd
import numpy as np
import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

# preprocessing , tworzenie df % chorob wg wieku [co 5 lat]


```python
# list_of_files = glob.glob('*.csv')
list_of_files = ['WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Respiratory diseases_13th January 2024 17_35.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Digestive diseases_13th January 2024 17_36.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Respiratory infections_13th January 2024 17_35.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Other neoplasms_13th January 2024 17_34.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Neuropsychiatric conditions_13th January 2024 17_37.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Malignant neoplasms_13th January 2024 17_33.csv', 'WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Cardiovascular diseases_13th January 2024 19_08.csv', 'WHOMortalityDatabase_Deaths_sex_age_a_country_area_year-Injuries_14th January 2024 11_28.csv']
print(list_of_files)
```

```python
columns = 'Region Code,Region Name,Country Code,Country Name,Year,Sex,Age group code,Age Group,Number,Percentage of cause-specific deaths out of total deaths,"Age-standardized death rate per 100 000 standard population","Death rate per 100 000 population"'.split(',')
print(columns)
```

```python
df = pd.DataFrame()
for file_name in list_of_files:
  new_df = pd.read_csv(file_name,skiprows=6,index_col=False)
  new_df["file_name"] = file_name
  df_merged = pd.concat([df, new_df],ignore_index=True)
  df = df_merged

df.head(1000)
```

```python
df.drop(columns=['Region Code','Region Name','Country Code','Age group code','Sex','Age-standardized death rate per 100 000 standard population','Death rate per 100 000 population','Percentage of cause-specific deaths out of total deaths'],inplace=True)
df.head()
```

```python
df.file_name.value_counts()
```

```python
df.rename(columns={'file_name':'Reason_of_death','Number':'Number_of_deaths'},inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Respiratory diseases_13th January 2024 17_35.csv','choroby i infekcje dróg oddechowych',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Respiratory infections_13th January 2024 17_35.csv','choroby i infekcje dróg oddechowych',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Other neoplasms_13th January 2024 17_34.csv','nowotwory',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Malignant neoplasms_13th January 2024 17_33.csv','nowotwory',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Digestive diseases_13th January 2024 17_36.csv','choroby układu trawienia',inplace=True)
df.replace('WHOMortalityDatabase_Deaths_sex_age_a_country_area_year-Injuries_14th January 2024 11_28.csv','wypadki',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Neuropsychiatric conditions_13th January 2024 17_37.csv','zaburzenia neuropsychiatryczne',inplace=True)
df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Cardiovascular diseases_13th January 2024 19_08.csv','choroby układu krwionośnego',inplace=True)
# df.replace('WHOMortalityDatabase_Trends_years_ages_a_country_area_sex-Unintentional injuries_13th January 2024 17_38.csv','wypadki',inplace=True)
df.head()
```

```python
l1 = list(df.loc[df['Reason_of_death'] == 'nowotwory']['Number_of_deaths'])[:19]
l2 = list(df.loc[df['Reason_of_death'] == 'nowotwory']['Number_of_deaths'])[19:]
deaths_nowotwory_combined = list(np.add(l1,l2))


df_nowotwory = df.loc[df['Reason_of_death'] == 'nowotwory'][:19]
df_nowotwory['Number_of_deaths'] = deaths_nowotwory_combined

df = pd.concat([df.loc[df['Reason_of_death'] != 'nowotwory'],df_nowotwory]).reset_index(drop=True)
df
```

```python
l1 = list(df.loc[df['Reason_of_death'] == 'choroby i infekcje dróg oddechowych']['Number_of_deaths'])[:19]
l2 = list(df.loc[df['Reason_of_death'] == 'choroby i infekcje dróg oddechowych']['Number_of_deaths'])[19:]
deaths_n_combined = list(np.add(l1,l2))


df_n = df.loc[df['Reason_of_death'] == 'choroby i infekcje dróg oddechowych'][:19]
df_n['Number_of_deaths'] = deaths_n_combined

df = pd.concat([df.loc[df['Reason_of_death'] != 'choroby i infekcje dróg oddechowych'],df_n]).reset_index(drop=True)
df
```

```python
df1 = pd.DataFrame(df.values,columns=['Kraj','Rok','Grupa wiekowa','Liczba zgonów','Przyczyna zgonu'])
```

```python
new_df = pd.read_csv('/content/wszystkie zgony.csv',skiprows=6,index_col=False)
total_deaths_per_age = new_df.Number[:-1]
print(total_deaths_per_age)
```

```python
lits_of_total_deaths = []
l = list(total_deaths_per_age)
for i in range(6):
  lits_of_total_deaths.extend(l)

wszystkie_zgony_df = pd.DataFrame(lits_of_total_deaths,columns=["Wszystkie zgony"])

df1['Procent zgonów'] = ((df1['Liczba zgonów'] / wszystkie_zgony_df['Wszystkie zgony']) * 100).astype('float').round(2)
df_co5 = df1
df1.to_csv('df_przyczyny_zgonow_kat_wiek_co5lat.csv',encoding='utf-16')
```

#tworzenie df % chorob wg wieku [20- 39] i [40 +]


```python
plt.rcParams['figure.figsize'] = [20,5]
sns.catplot(data=df1, x="Grupa wiekowa", y="Procent zgonów",jitter=False)
plt.xticks(rotation=90)
plt.show()
```

tutaj zostaje podjeta decyzja o podziale na [20-39] i [40+] , widać punkt zwrotny w danych w okolicach [40 -44]

```python
df1.replace(['[0]','[1-4]','[5-9]','[10-14]','[15-19]'],'[0-19]',inplace=True)
df1.replace(['[20-24]' ,'[25-29]','[30-34]','[35-39]'],'[20-39]',inplace=True)
df1.replace(['[40-44]', '[45-49]', '[50-54]' ,'[55-59]' ,'[60-64]' ,'[65-69]' ,'[70-74]' ,'[75-79]', '[80-84]' ,'[85+]'],'[40+]',inplace=True)
df1 = df1.loc[df1['Grupa wiekowa'] != '[0-19]'].reset_index(drop=True)
df1.head(20)
```

```python
sum40m = total_deaths_per_age[5:9].sum()
sum40w = total_deaths_per_age[9:].sum()

list_of_total_deaths_wide_age_range = []
for i in range(6):
  list_of_total_deaths_wide_age_range.append(sum40m)
  list_of_total_deaths_wide_age_range.append(sum40w)

print(list_of_total_deaths_wide_age_range)
```

```python
df_wide_age_group = pd.DataFrame(columns=df1.columns[:-1])
for przyczyna in df1['Przyczyna zgonu'].unique():
  g40m,g40w = df1.loc[df1['Przyczyna zgonu'] == przyczyna].groupby('Grupa wiekowa')['Liczba zgonów'].sum().values
  new_row = {'Kraj':'Poland','Rok':2020,'Grupa wiekowa': '[20-39]','Liczba zgonów':g40m,'Przyczyna zgonu':przyczyna}
  df_wide_age_group = df_wide_age_group.append(new_row,ignore_index=True)
  new_row = {'Kraj':'Poland','Rok':2020,'Grupa wiekowa': '[40 +]','Liczba zgonów':g40w,'Przyczyna zgonu':przyczyna}
  df_wide_age_group = df_wide_age_group.append(new_row,ignore_index=True)

df_wide_age_group['Procent zgonów'] = ((df_wide_age_group['Liczba zgonów'] / list_of_total_deaths_wide_age_range) * 100).round(2)
df_wide_age_group.to_csv('df_przyczyny_zgonow_kat_wiek_2039i40plus.csv',encoding='utf-16')
```
